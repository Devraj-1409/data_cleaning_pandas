# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\valmi\Downloads\Customer Call List.xlsx")
df
```

```python
df
```

```python
df.drop_duplicates()
```

```python
df['Last_Name']=df['Last_Name'].str.strip('/_...')
df.drop(columns = 'Not_Useful_Column',inplace = True)
df
```

```python
df['Phone_Number'] =df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex= True)
df
```

```python
df['Phone_Number']= df['Phone_Number'].astype('string')
```

```python
print(df.dtypes)
```

```python
df['Phone_Number']= df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('Na--','')
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('nan--','') 
df
df['Last_Name']= df['Last_Name'].fillna('Scott')
df
```

```python
df[['Street_Address','State','ZipCode']] =df['Address'].str.split(',',expand = True)
df
```

```python
df.drop(columns = 'Address')
```

```python
df['Paying Customer']= df['Paying Customer'].str.replace('Yes',"Y")
df['Paying Customer']= df['Paying Customer'].str.replace('No',"N")
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes',"Y")
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No',"N")
df
```

```python
df =df.fillna('')
```

```python
df
```

```python
df = df.replace('N/a','')
df
```

```python
for row in df.index:
    if df.loc[row,'Do_Not_Contact'] == "Y": 
        df.drop(row,inplace = True) 
```

```python
df
```

```python
for row in df.index:
    if df.loc[row,'Phone_Number'] == "": 
        df.drop(row,inplace = True)
        
```

```python
df
```

```python
df = df.drop(columns = "ZipCode")
```

```python
df = df.reset_index(drop = True)
```
